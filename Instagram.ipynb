```python
#Рассмотрим Петроградский район
#[широта, долгота], [параллель, меридиан] или [latitude, longtitude]
apteknab=[59.978620,30.315648]#верхняя граница Петроградского района - Аптекарская набережная
zaya=[59.948175,30.316601]#нижняя граница - Заячий остров
petrplos=[59.962036,30.253173]#левая - Петровская площадь (не берём Петровскую косу)
petrogrnab=[59.962639,30.33454]#правая - Петроградская набережная

#Найдём расстояние в градусах между верхней и нижней, левой и правой границами
vert=apteknab[0]-zaya[0]
goriz=petrogrnab[1]-petrplos[1]
#print(goriz,vert)

#В Яндекс.Картах горизонтальное расстояние между крайними точками приблизительно равно 4500 м, вертикальное - 3450
#В Инстаграме поиск соседних локаций происходит на радиусе около 500 м, поэтому посмотрим, сколько раз нам нужно будет поменять
#координаты центра, сдвигая "кружочек"
razmer=[round(4500/500),round(3450/500)]
#print(razmer)
#Получили, что по горизонтали достаточно сделать девять шагов, а по вертикали семь

#А теперь всё наоборот: чем больше ШИРОТА, тем ВЫШЕ точка; чем больше ДОЛГОТА, тем ПРАВЕЕ точка
#Поэтому начнём с нижнего левого угла Петроградского района: нужно взять самую НИЗКУЮ ШИРОТУ и самую ЛЕВУЮ ДОЛГОТУ
#Это будет ШИРОТА Заячьего острова (нижней границы) и ДОЛГОТА Петровской площади (левой границы)
nizleft=[zaya[0],petrplos[1]]
#Заполним матрицу координат: каждый раз нужно сдвигать ШИРОТУ на 1/7 расстояния, а ДОЛГОТУ на 1/9
#Не забываем, что делений 7 и 9, а значит, точек 8 и 10!
#Создаём пустой массив 8х10, куда будут добавляться координаты
matr=[[0 for i in range(razmer[0]+1)]for j in range(razmer[1]+1)]
for i in range(len(matr)):
    for j in range(len(matr[i])):
        matr[i][j]=[nizleft[0]+vert/razmer[1]*i,nizleft[1]+goriz/razmer[0]*j]
#print(matr)#наборы всех возможных центров, удалённых друг от друга на 500 м по горизонтали и вертикали

#Мы получили "сетку", охватывающую прямоугольник чуть больше, чем Петроградский район. Для удобства сделаем одноуровневый 
#Flatten, которого, конечно, здесь нет (сделаем из списка списков 8х10 один список длиной 80)
coords=[]
for i in range(len(matr)):
    for j in range(len(matr[i])):
        coords.append(matr[i][j])
print(coords)
```

```python
#Теперь необходимо вставить все координаты в API Инстаграма, чтобы собрать информацию о локациях в радиусе 
#около 500 м вокруг каждого из этих центров
#Импортируем библиотеки, необходимые для открытия сайтов
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
#Наш URL выглядит так: 
#https://api.instagram.com/v1/locations/search?lat=59.9390&lng=30.3158&access_token=1502967621.9c7bfcf.a5355e81b9244c31a7276a9515bda089
#Вместо lat нужно вставлять широту, вместо lng - долготу. Разобьём строку на пять частей:
api1="https://api.instagram.com/v1/locations/search?lat="
#после api1 - будущая широта
api2="&lng="
#здесь будущая долгота
api3="&access_token=1502967621.9c7bfcf.a5355e81b9244c31a7276a9515bda089"
#Создадим список URL-ов длиной 80, меняя координаты:
allurls=[api1+str(coords[i][0])+api2+str(coords[i][1])+api3 for i in range(len(coords))]
print(allurls)
```

```python
#Далее нам понадобится функция, разбивающая список а на сколько-то частей длины size:
def splitby(a, size):
    arr=[]
    while len(a)>size:
        p=a[:size]
        arr.append(p)
        a=a[size:]
    arr.append(a)
    return arr
```

```python
#Сделаем для всех пар координат, взяв их в качестве центра: напишем функцию, обрабатывающую весь массив по Петроградскому району
def locids(url1):
    html=urlopen(url1)
    soup=BeautifulSoup(html,'html.parser')
    #print(soup.prettify())#вся информация, которую выдаёт сайт, в формате BeautifulSoup
    #Превращаем это в строку
    s=str(soup)[10:len(str(soup))-25]#убираем лишнее в начале и в конце
    #print(s)

    #А теперь превратим нашу огромную строку в список (потому что уж очень она на него похожа внешне)
    s1=s.replace("{","").replace("}","")#уберём фигурные скобки из нашей большой-большой строки
    #print(s1)
    s2=s1.split('"id":')#разобьём на строки так, чтобы доказательством начала каждой из них служило слово '"id":', оно исчезнет
    #print(s2)
    s3=splitby(s2,1)[1:]#засунем каждую полученную строку в список (по факту - нарисуем квадратные скобки) и уберём следы первого id
    #print(s3)
    s4=[s3[i][0].split('",') for i in range(len(s3))]#у нас есть список списков, и каждый внутренний содержит в себе одну строку,
    #поэтому разобьём каждую строку на несколько строк, чтобы индикатором служили закрывающие кавычки и запятая
    #print(s4)
    #s5=[[int(s4[i][0][2:]), s4[i][1][10:], float(s4[i][2].split(",")[0][13:]), float(s4[i][2].split(",")[1][14:])] for i in range(len(s4))]
    #получаем список списков вида [целочисленный id, название локации, широта, долгота]
    
    #Если название локации нужно удалить, то просто в s5 не писать s4[i][1][10:], а также широту и долготу
    s5=[int(s4[i][0][2:]) for i in range(len(s4))]
    return s5
    #Подробно:
    #мы берём int(s4[i][0][2:]) - чтобы id начинался не с апострофа и не с пробела и был целочисленным
    #s4[i][1][10:] - название локации без слова "name" и прочих лишних символов
    #float(s4[i][2].split(",")[0][13:]) - широта и долгота записаны в s4 вместе, поэтому разбиваем его по индикатору-запятой,
    #получаем список из строки с широтой и строки с долготой, для широты (нулевой элемент) берём с 13-го символа вещественное значение
    #для долготы аналогично
```

```python
def kolvopub(idlocatsii):
    html1=urlopen(idlocatsii)
    soup1=BeautifulSoup(html1,'html.parser')
    soup2=str(soup1)
    #название локации
    locname=soup2[soup2.find('\n<title>\n')+len('\n<title>\n'):soup2.find(' on Instagram')].replace("amp;","")
    #количество публикаций в локации
    kolvo=int(soup2[soup2.find('"edge_location_to_media":{"count":')+len('"edge_location_to_media":{"count":'):soup2.find(',"page_info"')])
    return locname, kolvo
```
