```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def latlong(url):
    html=urlopen(url)
    soup=BeautifulSoup(html,'html.parser')
    s=str(soup)
    lat=s[s.find('property="og:url">\n<meta content="')+len('property="og:url">\n<meta content="'):s.find('" property="place:location:latitude">\n<meta content="')]
    long=s[s.find('property="place:location:latitude">\n<meta content="')+len('property="place:location:latitude">\n<meta content="'):s.find('" property="place:location:longitude"/>\n<meta content')]
    return [lat,long]
```

```python
f1=open('C://Users/forte/Desktop/ЯиМП/Инстаграм/Итоги.csv','r')
a=[]
for line in f1:
    a.append(line.replace("\n","").split(";"))
f1.close()
print(a)
```

```python
b=[]
for i in range(len(a)):
    if i==0:
        b.append(["Широта","Долгота"])
    else:
        b.append(latlong(a[i][2]))
b
```

```python
for i in range(len(a)):
    a[i].append(b[i][0])
    a[i].append(b[i][1])
a
```

```python
f2=open('C://Users/forte/Desktop/ЯиМП/Инстаграм/ИтогиШД.csv','w')#запись таблицы в csv-файл
for i in range(len(a)):
    f2.write(';'.join(a[i])+'\n')
f2.close()
```
